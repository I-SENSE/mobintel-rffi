```python
def read_dataset(file_path):
    # Open the HDF5 file
    with h5py.File(file_path, 'r') as h5_file:
        # Extract the dataset named 'data'
        data = h5_file['data'][:]
        labels = h5_file['label'][:] 

    print(labels.shape)
    print(data.shape)
    return [data, labels]

def analyze_labels(labels):
    devices = defaultdict(int)

    for i in np.arange(0, labels.shape[0]):
        key = str(labels[i])
        devices[key] = devices[key] + 1

    keys = list(devices.keys())
    print(len(keys))
    keys.sort()
    print(keys)
    return devices
```

```python
[data_train, labels_train] = read_dataset(dataset_train)
[data_enrol, labels_enrol] = read_dataset(dataset_enrol)
[data_identify, labels_identify] = read_dataset(dataset_identify)
```
